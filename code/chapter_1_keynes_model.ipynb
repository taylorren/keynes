# 《通论》第一章：前言和性质的探讨

本笔记本展示了凯恩斯在《就业、利息和货币通论》第一章中提出的核心思想，特别是古典经济学与凯恩斯经济学之间的根本差异。

```python
# 导入必要的库
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
from IPython.display import display, Math, Markdown
%matplotlib inline

# 设置LaTeX和中文字体支持
rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial', 'Times New Roman']
rcParams['font.family'] = 'sans-serif'
rcParams['mathtext.fontset'] = 'stix'  # 使用STIX字体支持数学公式
rcParams['axes.unicode_minus'] = False
```

## 1. 定义核心经济变量

首先，我们定义凯恩斯在《通论》中使用的核心经济变量：

```python
# 定义符号变量
Y = sp.Symbol('Y', real=True, positive=True)  # 总产出/收入
C = sp.Symbol('C', real=True, positive=True)  # 消费
I = sp.Symbol('I', real=True, positive=True)  # 投资
L = sp.Symbol('L', real=True, positive=True)  # 货币需求
M = sp.Symbol('M', real=True, positive=True)  # 货币供给
r = sp.Symbol('r', real=True)                 # 利率
N = sp.Symbol('N', real=True, positive=True)  # 就业量
P = sp.Symbol('P', real=True, positive=True)  # 物价水平
Z = sp.Symbol('Z', real=True, positive=True)  # 总供给价格
D = sp.Symbol('D', real=True, positive=True)  # 总需求价格
E = sp.Symbol('E', real=True, positive=True)  # 总支出 (aggregate expenditure)

print("定义的核心经济变量：")
print(f"总产出/收入 (Y): {Y}")
print(f"消费 (C): {C}")
print(f"投资 (I): {I}")
print(f"总需求价格 (D): {D}")
print(f"总供给价格 (Z): {Z}")
```

## 2. 古典经济学假设：萨伊定律

古典经济学的核心假设是萨伊定律："供给创造其自身的需求"。这意味着总供给总是等于总需求。

```python
print("--- 古典经济学假设：萨伊定律 ---")
print("核心观点：供给创造其自身的需求")
print("数学表达：总供给 = 总需求 (Z = D)")

# 萨伊定律方程
say_law = sp.Eq(Z, D)
print(f"萨伊定律方程: {say_law}")
display(Math(sp.latex(say_law)))

# 古典经济学中的充分就业假设
print("\n古典经济学中的充分就业假设:")
full_employment_condition = sp.Eq(Y, Z)
print(f"充分就业条件: {full_employment_condition}")
display(Math(sp.latex(full_employment_condition)))
print("即：总收入等于总供给价格，意味着充分就业")
```

## 3. 凯恩斯的挑战：有效需求原理

凯恩斯指出，市场可能长期处于非充分就业均衡状态，因为有效需求可能不足。

```python
print("--- 凯恩斯的挑战：有效需求原理 ---")
print("核心观点：总需求可能不足，导致非自愿失业")
print("数学表达：总需求可能小于总供给 (D < Z)")

# 定义消费函数 (凯恩斯的心理消费倾向)
c_0 = sp.Symbol('c_0', real=True, positive=True)  # 自发消费
mpc = sp.Symbol('c', real=True, positive=True)   # 边际消费倾向 (用c代替mpc避免冲突)

# 消费函数: C = c_0 + mpc * Y
consumption_function = c_0 + mpc * Y
print(f"\n消费函数: C = {consumption_function}")
display(Math(sp.latex(consumption_function)))
print("其中 c_0 是自发消费，c 是边际消费倾向")

# 总需求函数 (简化版): D = C + I
I_bar = sp.Symbol('I_0', positive=True)  # 自主投资
aggregate_demand = consumption_function + I_bar
print(f"\n总需求函数: D = {aggregate_demand}")
display(Math(sp.latex(aggregate_demand)))
# 将总支出符号 E 与总需求函数关联并显示
display(Math(sp.latex(sp.Eq(E, aggregate_demand))))
```

## 4. 均衡条件对比

古典经济学和凯恩斯经济学对均衡的理解存在根本差异。

```python
print("--- 均衡条件对比 ---")
classical_equilibrium = sp.Eq(D, Z)  # 古典：总需求 = 总供给
keynesian_equilibrium = sp.Eq(D, Y)  # 凯恩斯：总需求 = 总收入

print(f"古典均衡条件: {classical_equilibrium}")
display(Math(sp.latex(classical_equilibrium)))
print(f"凯恩斯均衡条件: {keynesian_equilibrium}")
display(Math(sp.latex(keynesian_equilibrium)))

# 非自愿失业的概念
print("\n--- 非自愿失业的概念 ---")
print("当 D < Z 时，存在总需求不足，导致非自愿失业")
unemployment_condition = sp.Lt(D, Z)
print(f"非自愿失业条件: {unemployment_condition}")
display(Math(sp.latex(unemployment_condition)))
```

## 5. 数值模拟与可视化

现在我们设定具体参数值，进行数值模拟并可视化结果。

```python
# 设定参数值
params = {
    c_0: 100,      # 自发消费
    mpc: 0.8,      # 边际消费倾向
    I_bar: 200,    # 自主投资
}

# 计算均衡收入
equilibrium_eq = sp.Eq(Y, aggregate_demand.subs(params))
equilibrium_income = sp.solve(equilibrium_eq, Y)[0]
print(f"\n均衡收入: Y* = {equilibrium_income}")
display(Markdown(f"$Y^* = {sp.latex(equilibrium_income)} = {float(equilibrium_income):.2f}$"))

# 计算均衡消费
equilibrium_consumption = consumption_function.subs({**params, Y: equilibrium_income})
print(f"均衡消费: C* = {equilibrium_consumption}")
display(Markdown(f"$C^* = {sp.latex(equilibrium_consumption)} = {float(equilibrium_consumption):.2f}$"))

# 计算均衡总需求
equilibrium_ad = aggregate_demand.subs({**params, Y: equilibrium_income})
print(f"均衡总需求: D* = {equilibrium_ad}")
display(Markdown(f"$D^* = {sp.latex(equilibrium_ad)} = {float(equilibrium_ad):.2f}$"))
```

```python
# 可视化凯恩斯交叉模型
plt.figure(figsize=(12, 8))

# 创建数值范围，确保包含均衡收入以观察乘数效应
upper = max(1000, float(equilibrium_income) * 1.1)
Y_range = np.linspace(0, upper, 400)
C_range = [float(consumption_function.subs({**params, Y: val})) for val in Y_range]
AD_range = [float(aggregate_demand.subs({**params, Y: val})) for val in Y_range]

# 子图1：凯恩斯交叉图
plt.subplot(2, 1, 1)
plt.plot(Y_range, Y_range, 'k--', label=r'$Y = E$', linewidth=2)
plt.plot(Y_range, AD_range, 'b-', label=rf'$AD = {sp.latex(aggregate_demand.subs(params))}$', linewidth=2)
plt.axvline(x=float(equilibrium_income), color='r', linestyle=':', label=rf'$Y^* = {float(equilibrium_income):.2f}$')
plt.axhline(y=float(equilibrium_income), color='r', linestyle=':', label=rf'$E^* = {float(equilibrium_income):.2f}$')

plt.xlabel(r'总收入 $Y$')
plt.ylabel(r'总支出 $E$')
plt.title(r'凯恩斯交叉模型 - 《通论》第一章')
plt.legend()
plt.grid(True, alpha=0.3)

# 子图2：消费函数
S = Y - C  # 储蓄等于收入减去消费
saving_func = Y - consumption_function
saving_range = [float(saving_func.subs({**params, Y: val})) for val in Y_range]

plt.subplot(2, 1, 2)
plt.plot(Y_range, C_range, 'g-', label=rf'$C = {sp.latex(consumption_function.subs({c_0: params[c_0], mpc: params[mpc]}))}$', linewidth=2)
plt.plot(Y_range, saving_range, 'orange', label=rf'$S = Y - C = Y - {sp.latex(consumption_function.subs({c_0: params[c_0], mpc: params[mpc]}))}$', linewidth=2)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel(r'总收入 $Y$')
plt.ylabel(r'消费 $C$ 和 储蓄 $S$')
plt.title(r'消费与储蓄函数')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. 乘数效应

边际消费倾向决定了乘数效应的大小。

```python
# 计算乘数
multiplier = 1 / (1 - mpc)
print(f"乘数公式: k = {multiplier}")
display(Math(sp.latex(multiplier)))

# 代入数值
multiplier_value = 1 / (1 - params[mpc])
print(f"数值乘数: k = {multiplier_value:.2f}")

print(f"\n乘数效应含义：")
print(f"当自主支出增加1单位时，总收入将增加{multiplier_value:.2f}单位")
print(f"这是因为增加的收入会进一步刺激消费，形成连锁反应")
```

## 7. 边际消费倾向的影响

让我们看看不同的边际消费倾向如何影响均衡收入。

```python
# 不同边际消费倾向的影响
plt.figure(figsize=(10, 6))
# 创建数值范围，确保包含均衡收入以观察不同MPC下的均衡点
upper = max(1000, float(equilibrium_income) * 1.1)
Y_range = np.linspace(0, upper, 400)

mpcs = [0.6, 0.7, 0.8, 0.9]
colors = ['blue', 'green', 'red', 'purple']

for i, mpc_val in enumerate(mpcs):
    params_temp = {c_0: 100, mpc: mpc_val, I_bar: 200}
    AD_temp = [float(aggregate_demand.subs({**params_temp, Y: val})) for val in Y_range]
    plt.plot(Y_range, AD_temp, label=f'MPC = {mpc_val}, Multiplier = {1/(1-mpc_val):.2f}', linewidth=2, color=colors[i])

plt.plot(Y_range, Y_range, 'k--', label='45°线', linewidth=2)
plt.xlabel('总收入 (Y)')
plt.ylabel('总支出')
plt.title('不同边际消费倾向对均衡收入的影响')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 计算不同MPC下的均衡收入
print("不同边际消费倾向下的均衡收入：")
for mpc_val in mpcs:
    params_temp = {c_0: 100, mpc: mpc_val, I_bar: 200}
    eq_income = sp.solve(sp.Eq(Y, aggregate_demand.subs(params_temp)), Y)[0]
    print(f"MPC = {mpc_val}: 均衡收入 Y* = {eq_income} = {float(eq_income):.2f}, 乘数 = {1/(1-mpc_val):.2f}")
```

```python
# 对初始自主支出 (I_bar) 冲击进行敏感性分析，展示乘数放大效应并标注 ΔY
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
# 选择更明显的冲击值以便观察放大效应
I_bars = [100, 200, 500]
colors = ['blue', 'green', 'red']
# 重新计算Y范围，确保包含所有均衡点
upper = max(1000, float(equilibrium_income) * 1.5)
Y_range = np.linspace(0, upper, 400)
eq_values = []
for i, I_val in enumerate(I_bars):
    params_temp = {c_0: params[c_0], mpc: params[mpc], I_bar: I_val}
    AD_temp = [float(aggregate_demand.subs({**params_temp, Y: val})) for val in Y_range]
    plt.plot(Y_range, AD_temp, label=rf'$AD\ (I_0={I_val})$', color=colors[i], linewidth=2)
    eq_y = sp.solve(sp.Eq(Y, aggregate_demand.subs(params_temp)), Y)[0]
    eq_values.append(float(eq_y))
    plt.axvline(x=float(eq_y), color=colors[i], linestyle='--', alpha=0.8)
    plt.scatter([float(eq_y)], [float(eq_y)], color=colors[i])
    plt.text(float(eq_y)*1.01, float(eq_y)*0.95, f'Y*={float(eq_y):.1f}', color=colors[i])
# 画45°线
plt.plot(Y_range, Y_range, 'k--', label='45°线')
plt.xlabel('总收入 (Y)')
plt.ylabel('总支出 E')
plt.title('自主支出冲击下的凯恩斯交叉：展示乘数放大效应')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
# 显示数值结果并标注ΔY（相对于第一个基准I_bar）
baseline = eq_values[0]
from IPython.display import Markdown, display
rows = ['| I_bar | Y* | ΔY (相对 I_bar='+str(I_bars[0])+') |', '|---:|---:|---:|']
for I_val, y_val in zip(I_bars, eq_values):
    delta = y_val - baseline
    rows.append(f'| {I_val} | {y_val:.2f} | {delta:.2f} |')
display(Markdown('\n'.join(rows)))
```

## 8. 主要结论

通过以上分析，我们可以得出以下重要结论：

1. **古典经济学**认为市场机制能够自动调节，使经济达到充分就业的均衡状态。
2. **凯恩斯经济学**指出，由于有效需求不足，经济可能长期处于非充分就业均衡状态。
3. **政府干预**可以通过增加总需求来提高就业水平。
4. **边际消费倾向**影响乘数效应的大小，从而影响政策效果。
5. **乘数效应**解释了为什么初始的支出增加会导致总收入的更大增长。

这些概念构成了凯恩斯宏观经济理论的基础，为后续章节的深入分析奠定了基础。
